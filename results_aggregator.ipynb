### PAD

```python
import os
import pandas as pd

micro_res = pd.DataFrame()
macro_res = pd.DataFrame()
weighted_res = pd.DataFrame()
results = [micro_res, macro_res, weighted_res]
res_dir = '.'
metrics = ['micro avg', 'macro avg', 'weighted avg']
for subdir, dirs, files in os.walk(res_dir):
    for filename in files:
        name = ''
#         if filename.find('W2V') == -1:
#             continue
        if filename.find('results_aggregator') != -1:
            continue
        if filename.find('RFQ') != -1:
            continue
        if subdir.find('.git') != -1:
            continue
        with open(os.path.join(subdir, filename)) as f:
            mode = 0
            for line in f:
                if line.find('"cell_type": "code"') != -1:
                    mode = 0
                elif line.find('"cell_type": "markdown"') != -1:
                    mode = 1
                line = line.strip().strip('"')
                if mode and line.find('##') != -1:
                    name = line.strip('#').strip()
            
                for metric, res in zip(metrics, results):
                    if line.find(metric) != -1:
                        line = line.strip('"')
                        if name == "":
                            res.at[filename.strip('.ipynb'), 'precision'] = line.split()[2]
                            res.at[filename.strip('.ipynb'), 'recall'] = line.split()[3]
                            res.at[filename.strip('.ipynb'), 'f1-score'] = line.split()[4]
                        else:
                            res.at[name, 'precision'] = line.split()[2]
                            res.at[name, 'recall'] = line.split()[3]
                            res.at[name, 'f1-score'] = line.split()[4]
```

LaTex format

```python
for index, row in micro_res.sort_values('f1-score', ascending=False).iterrows():
    print('{} & {} & {} & {} \\\\'.format(index.replace('_', '\\_'), row['precision'], row['recall'], row['f1-score']))
```

```python
micro_res.sort_values('f1-score', ascending=False)
```
